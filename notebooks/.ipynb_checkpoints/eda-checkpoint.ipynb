## I. Importing Essential Libraries

```python
# Arrays and dataframes
import numpy as np
import pandas as pd
import scipy as sp

# Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

## II. Loading Preprocessed Data

```python
train_dataset = pd.read_csv("../data/preprocessed/Unsampled_train_dataset.csv")
train_dataset_undersampled = pd.read_csv("../data/preprocessed/Undersampled_train_dataset.csv")
train_dataset_oversampled = pd.read_csv("../data/preprocessed/Oversampled_train_dataset.csv")
train_dataset_smote = pd.read_csv("../data/preprocessed/SMOTE_train_dataset.csv")

train_datasets_dict = {"Unsampled": train_dataset, "Undersampled": train_dataset_undersampled, "Oversampled": train_dataset_oversampled, "SMOTE": train_dataset_smote}

unsampled = train_datasets_dict["Unsampled"]
undersampled = train_datasets_dict["Undersampled"]
oversampled = train_datasets_dict["Oversampled"]
smote = train_datasets_dict["SMOTE"]
```

## IV. Feature Skewness

```python
for name, data in train_datasets_dict.items():
    plt.figure(figsize = (20, 10))
    plt.title("{} Skewness".format(name))
    sns.barplot(data.columns, sp.stats.skew(data))
    plt.show()
```

## V. Feature-Label Boxplots

```python
NUM_BINS = 500
```

```python
def remove_outliers(df, feature_name):
    return df[np.abs(df[feature_name] - df[feature_name].mean()) <= (3 * df[feature_name].std())][feature_name]
```

```python
# Unsampled histograms
fig, ax = plt.subplots(nrows = 4, figsize = (20, 20))
fig.suptitle("Unsampled Skewed Features")

# v8 histogram
ax[0].set_title("v8 vs label")
unsampled_v8_no_outliers = remove_outliers(unsampled, "v8")
sns.distplot(unsampled_v8_no_outliers, bins = NUM_BINS, ax = ax[0])

# v23 histogram
unsampled_v23_no_outliers = remove_outliers(unsampled, "v23")
ax[1].set_title("v23 vs label")
sns.distplot(unsampled_v23_no_outliers, bins = NUM_BINS, ax = ax[1])

# v28 histogram
unsampled_v28_no_outliers = remove_outliers(unsampled, "v28")
ax[2].set_title("v28 vs label")
sns.distplot(unsampled_v28_no_outliers, bins = NUM_BINS, ax = ax[2])

# amount histogram
unsampled_amount_no_outliers = remove_outliers(unsampled, "amount")
ax[3].set_title("amount vs label")
sns.distplot(unsampled_amount_no_outliers, bins = NUM_BINS, ax = ax[3])

plt.show()
```

```python
# Undersampled histograms
fig, ax = plt.subplots(nrows = 4, figsize = (20, 20))
fig.suptitle("Undersampled Skewed Features")

# v7 histogram
ax[0].set_title("v7 vs label")
undersampled_v7_no_outliers = remove_outliers(undersampled, "v7")
sns.distplot(undersampled_v7_no_outliers, bins = NUM_BINS, ax = ax[0])

# v21 histogram
undersampled_v21_no_outliers = remove_outliers(undersampled, "v21")
ax[1].set_title("v21 vs label")
sns.distplot(undersampled_v21_no_outliers, bins = NUM_BINS, ax = ax[1])

# v22 histogram
undersampled_v22_no_outliers = remove_outliers(undersampled, "v22")
ax[2].set_title("v22 vs label")
sns.distplot(undersampled_v22_no_outliers, bins = NUM_BINS, ax = ax[2])

# amount histogram
undersampled_amount_no_outliers = remove_outliers(undersampled, "amount")
ax[3].set_title("amount vs label")
sns.distplot(undersampled_amount_no_outliers, bins = NUM_BINS, ax = ax[3])

plt.show()
```

```python
# Oversampled histograms
fig, ax = plt.subplots(nrows = 4, figsize = (20, 20))
fig.suptitle("Oversampled Skewed Features")

# v7 histogram
ax[0].set_title("v7 vs label")
oversampled_v7_no_outliers = remove_outliers(oversampled, "v7")
sns.distplot(oversampled_v7_no_outliers, bins = NUM_BINS, ax = ax[0])

# v21 histogram
oversampled_v21_no_outliers = remove_outliers(oversampled, "v21")
ax[1].set_title("v21 vs label")
sns.distplot(oversampled_v21_no_outliers, bins = NUM_BINS, ax = ax[1])

# v23 histogram
oversampled_v23_no_outliers = remove_outliers(oversampled, "v23")
ax[2].set_title("v23 vs label")
sns.distplot(oversampled_v23_no_outliers, bins = NUM_BINS, ax = ax[2])

# amount histogram
oversampled_amount_no_outliers = remove_outliers(oversampled, "amount")
ax[3].set_title("amount vs label")
sns.distplot(oversampled_amount_no_outliers, bins = NUM_BINS, ax = ax[3])

plt.show()
```

## VI. Bivariate Analysis

```python
# Feature-label correlation barplot
for name, data in train_datasets_dict.items():
    plt.figure(figsize=(20, 10))
    plt.title("{} Feature-Label Correlation".format(name))
    sns.barplot(data.drop("label", axis = 1).columns, data.corr()["label"].iloc[:-1])
    plt.show()
```

```python
# Feature-feature correlation heatmap
for name, data in train_datasets_dict.items():
    plt.figure(figsize=(20, 10))
    plt.title("{} Feature-Feature Correlation".format(name))
    sns.heatmap(data.corr())
    plt.show()
```

## VII. Outlier Detection

```python
for name, data in train_datasets_dict.items():
    # Find interquartile range
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1

    # Consolidate dataset indexes
    data.reset_index(drop = True, inplace = True)

    # Detect outliers
    outlier_index = data[(data < q1 - 1.5 * iqr) | (data > q3 + 1.5 * iqr)].dropna(how = "all").index
    outliers = data.iloc[outlier_index.to_list()]
    
    print("{} Outlier Count:\n{}\n".format(name, outliers["label"].value_counts()))
```
