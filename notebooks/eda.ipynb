## I. Importing Essential Libraries

```python
# Arrays and dataframes
import numpy as np
import pandas as pd
import scipy as sp

# Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

## II. Loading Preprocessed Data

```python
train_dataset = pd.read_csv("../data/preprocessed/Unsampled_train_dataset.csv")
train_dataset_undersampled = pd.read_csv("../data/preprocessed/Undersampled_train_dataset.csv")
train_dataset_oversampled = pd.read_csv("../data/preprocessed/Oversampled_train_dataset.csv")
train_dataset_smote = pd.read_csv("../data/preprocessed/SMOTE_train_dataset.csv")

train_datasets_dict = {"Unsampled": train_dataset, "Undersampled": train_dataset_undersampled, "Oversampled": train_dataset_oversampled, "SMOTE": train_dataset_smote}

unsampled = train_datasets_dict["Unsampled"]
undersampled = train_datasets_dict["Undersampled"]
oversampled = train_datasets_dict["Oversampled"]
smote = train_datasets_dict["SMOTE"]
```

## IV. Feature Skewness

```python
for name, data in train_datasets_dict.items():
    plt.figure(figsize = (20, 10))
    plt.title("{} Skewness".format(name))
    sns.barplot(data.columns, sp.stats.skew(data))
    plt.show()
```

## VI. Bivariate Analysis

```python
# Feature-label correlation barplot
for name, data in train_datasets_dict.items():
    plt.figure(figsize=(20, 10))
    plt.title("{} Feature-Label Correlation".format(name))
    sns.barplot(data.drop("label", axis = 1).columns, data.corr()["label"].iloc[:-1])
    plt.show()
```

```python
# Feature-feature correlation heatmap
for name, data in train_datasets_dict.items():
    plt.figure(figsize=(20, 10))
    plt.title("{} Feature-Feature Correlation".format(name))
    sns.heatmap(data.corr())
    plt.show()
```

## VII. Outlier Detection

```python
for name, data in train_datasets_dict.items():
    # Find interquartile range
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1

    # Consolidate dataset indexes
    data.reset_index(drop = True, inplace = True)

    # Detect outliers
    outlier_index = data[(data < q1 - 1.5 * iqr) | (data > q3 + 1.5 * iqr)].dropna(how = "all").index
    outliers = data.iloc[outlier_index.to_list()]
    
    print("{} Outlier Count:\n{}\n".format(name, outliers["label"].value_counts()))
```
