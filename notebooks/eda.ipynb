## V. Skewed Histograms

```python
NUM_BINS = 500
```

```python
def remove_outliers(df, feature_name):
    return df[np.abs(df[feature_name] - df[feature_name].mean()) <= (3 * df[feature_name].std())][feature_name]
```

```python
# Unsampled histograms
fig, ax = plt.subplots(nrows = 4, figsize = (20, 20))
fig.suptitle("Unsampled Skewed Features")

# v8 histogram
ax[0].set_title("v8 vs label")
unsampled_v8_no_outliers = remove_outliers(unsampled, "v8")
sns.distplot(unsampled_v8_no_outliers, bins = NUM_BINS, ax = ax[0])

# v23 histogram
unsampled_v23_no_outliers = remove_outliers(unsampled, "v23")
ax[1].set_title("v23 vs label")
sns.distplot(unsampled_v23_no_outliers, bins = NUM_BINS, ax = ax[1])

# v28 histogram
unsampled_v28_no_outliers = remove_outliers(unsampled, "v28")
ax[2].set_title("v28 vs label")
sns.distplot(unsampled_v28_no_outliers, bins = NUM_BINS, ax = ax[2])

# amount histogram
unsampled_amount_no_outliers = remove_outliers(unsampled, "amount")
ax[3].set_title("amount vs label")
sns.distplot(unsampled_amount_no_outliers, bins = NUM_BINS, ax = ax[3])

plt.show()
```

## VII. Outlier Detection

```python
for name, data in train_datasets_dict.items():
    # Find interquartile range
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1

    # Detect outliers
    outlier_index = data[(data < q1 - 1.5 * iqr) | (data > q3 + 1.5 * iqr)].dropna(how = "all").index
    outliers = data.iloc[outlier_index.to_list()]
    
    # Compute outlier_value_counts and outlier_value_props
    outlier_value_counts = outliers["label"].value_counts()
    outlier_value_props = outlier_value_counts / data["label"].value_counts()
    
    # Plot
    fig, ax = plt.subplots(ncols = 2, figsize = (20, 5))
    sns.barplot(x = outlier_value_counts.index, y = outlier_value_counts, ax = ax[0])
    ax[0].set_title("{} Outlier Value Counts".format(name))
    sns.barplot(x = outlier_value_counts.index, y = outlier_value_props, ax = ax[1])
    ax[1].set_title("{} Outlier Value Prop".format(name))
    plt.plot()
```
